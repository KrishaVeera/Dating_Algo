```python
# === STEP 1: Overlap & Variety across OKCupid / Hinge / Bumble / Tinder ===
from itertools import combinations
ID_COL = "user_id" 

tables = {
    "okcupid": okcupid,
    "hinge": hinge,
    "bumble": bumble,
    "tinder": tinder,
}

# 1) basic sets
sets = {name: set(df[ID_COL].tolist()) for name, df in tables.items()}
sizes = {name: len(s) for name, s in sets.items()}
topN_counts = pd.DataFrame({"app": list(sizes.keys()), "topN_count": list(sizes.values())}).sort_values("app").reset_index(drop=True)

# 2) pairwise Jaccard (similarity of membership, ignores rank)
apps = list(sets.keys())
jacc = pd.DataFrame(index=apps, columns=apps, dtype=float)
for a in apps:
    for b in apps:
        inter = len(sets[a] & sets[b])
        union = len(sets[a] | sets[b]) if sets[a] or sets[b] else 1
        jacc.loc[a, b] = inter / union
jaccard_matrix = jacc

# 3) unique (variety) — how many appear only in that app’s list
unique_counts = {a: len(sets[a] - set().union(*[sets[b] for b in apps if b != a])) for a in apps}
variety = pd.DataFrame({
    "app": apps,
    "unique_only": [unique_counts[a] for a in apps],
    "unique_ratio": [unique_counts[a] / sizes[a] if sizes[a] else 0.0 for a in apps],
}).sort_values(["unique_only","unique_ratio"], ascending=False).reset_index(drop=True)

# 4) overlaps: pairwise, triple, all-four
def combo_count(names):
    s = sets[names[0]]
    for n in names[1:]:
        s = s & sets[n]
    return len(s)

pair_rows = [{"group":" & ".join(pair), "count": combo_count(pair)} for pair in combinations(apps, 2)]
triple_rows = [{"group":" & ".join(tri), "count": combo_count(tri)} for tri in combinations(apps, 3)]
all_four = [{"group":" & ".join(apps), "count": combo_count(apps)}]
overlap_table = pd.DataFrame(pair_rows + triple_rows + all_four).sort_values(["count","group"], ascending=[False,True]).reset_index(drop=True)

# 5) show
print("Top-N sizes per app:")
display(topN_counts)

print("\nUnique (variety) per app — higher = more variety:")
display(variety)

print("\nPairwise Jaccard similarity (1 = identical members, 0 = disjoint):")
display(jaccard_matrix.style.format("{:.2f}"))

print("\nOverlap counts (pairwise, triple, all four):")
display(overlap_table)
```

```python
ID_COL = "user_id"  # change if needed

tables = {
    "okcupid": okcupid,
    "hinge": hinge,
    "bumble": bumble,
    "tinder": tinder,
    "cupid": cupid}

# 1) basic sets & sizes
sets  = {name: set(df[ID_COL].tolist()) for name, df in tables.items()}
sizes = {name: len(s) for name, s in sets.items()}
topN_counts = (pd.DataFrame({"app": list(sizes.keys()), "topN_count": list(sizes.values())})
                 .sort_values("app").reset_index(drop=True))
print("Top-N sizes per app:")
display(topN_counts)

# 2) pairwise Jaccard (membership similarity, ignores rank)
apps = list(sets.keys())
jacc = pd.DataFrame(index=apps, columns=apps, dtype=float)
for a in apps:
    for b in apps:
        inter = len(sets[a] & sets[b])
        union = len(sets[a] | sets[b]) if sets[a] or sets[b] else 1
        jacc.loc[a,b] = inter / union
print("\nPairwise Jaccard similarity (1 = identical members, 0 = disjoint):")
display(jacc.style.format("{:.2f}"))

# 3) unique (variety): elements only in that app
unique_counts = {a: len(sets[a] - set().union(*[sets[b] for b in apps if b != a])) for a in apps}
variety = (pd.DataFrame({
            "app": apps,
            "unique_only": [unique_counts[a] for a in apps],
            "unique_ratio": [unique_counts[a] / sizes[a] if sizes[a] else 0.0 for a in apps],
          })
          .sort_values(["unique_only","unique_ratio"], ascending=False)
          .reset_index(drop=True))
print("\nUnique (variety) per app — higher = more variety:")
display(variety)

# 4) overlaps for all combination sizes (2..N)
def combo_ids(names):
    s = sets[names[0]]
    for n in names[1:]:
        s = s & sets[n]
    return s

rows = []
for r in range(2, len(apps)+1):
    for combo in combinations(apps, r):
        ids = combo_ids(combo)
        rows.append({"group": " & ".join(combo), "k": r, "count": len(ids)})

overlap_table = (pd.DataFrame(rows)
                 .sort_values(["k","count","group"], ascending=[True, False, True])
                 .reset_index(drop=True))
print("\nOverlap counts by combination size (pairs, triples, …, all):")
display(overlap_table)

# Optional: show the exact intersection shared by ALL apps
all_shared_ids = combo_ids(apps)
print(f"\nCandidates shared by ALL apps ({len(all_shared_ids)}):")
display(pd.DataFrame({ID_COL: sorted(all_shared_ids)}).head(20))
```

```python
# === STEP 2: Rank agreement (generalized to n apps) ===
import pandas as pd
from itertools import combinations
from scipy.stats import spearmanr

ID_COL = "user_id"   # candidate identifier
RANK_COL = "rank"    # the rank column we added when saving each algo

# Example: core 4 only
tables = {"okcupid": okcupid, "hinge": hinge, "bumble": bumble, "tinder": tinder}

apps = list(tables.keys())

# 1) make clean mapping of app → (id → rank)
rank_maps = {app: dict(zip(df[ID_COL], df[RANK_COL])) for app, df in tables.items()}

# 2) pairwise Spearman correlations
corrs = pd.DataFrame(index=apps, columns=apps, dtype=float)
for a, b in combinations(apps, 2):
    common_ids = set(rank_maps[a].keys()) & set(rank_maps[b].keys())
    if len(common_ids) > 1:
        ra = [rank_maps[a][i] for i in common_ids]
        rb = [rank_maps[b][i] for i in common_ids]
        rho, _ = spearmanr(ra, rb)
    else:
        rho = float("nan")
    corrs.loc[a, b] = rho
    corrs.loc[b, a] = rho

# diagonals = 1.0
for a in apps:
    corrs.loc[a, a] = 1.0

print("Spearman rank correlations (overlapping candidates only):")
display(corrs.style.format("{:.2f}"))
```

```python
# === STEP 3: Fairness, Balance, Novelty metrics ===
import pandas as pd

ID_COL   = "user_id"
AGE_COL  = "age"
SEX_COL  = "sex"
LOC_COL  = "location"

apps = list(tables.keys())

metrics = []

for app, df in tables.items():
    df = df.copy()
    
    # 1) Gender balance (simple proportion of females vs males, adjust if nonbinary exists)
    gender_counts = df[SEX_COL].value_counts(normalize=True)
    female_ratio = gender_counts.get("f", 0.0)
    male_ratio   = gender_counts.get("m", 0.0)
    
    # 2) Age diversity (standard deviation or IQR)
    age_std = df[AGE_COL].std(skipna=True)
    age_range = df[AGE_COL].max() - df[AGE_COL].min()
    
    # 3) Location spread (# of unique cities / states / countries)
    unique_locs = df[LOC_COL].nunique(dropna=True)
    
    # 4) Novelty = proportion of candidates that *only* appear in this app
    unique_ids = set(df[ID_COL]) - set().union(*[set(tables[b][ID_COL]) for b in apps if b != app])
    novelty_ratio = len(unique_ids) / len(df) if len(df) else 0.0
    
    metrics.append({
        "app": app,
        "female_ratio": round(female_ratio, 2),
        "male_ratio": round(male_ratio, 2),
        "age_std": round(age_std, 2),
        "age_range": age_range,
        "unique_locations": unique_locs,
        "novelty_ratio": round(novelty_ratio, 2),
    })

metrics_df = pd.DataFrame(metrics).sort_values("app").reset_index(drop=True)

print("Step 3: Fairness / Balance / Novelty Metrics")
display(metrics_df)
```

```python
# ===User-centric metrics & trade-offs (fixed hinge col) ===
import pandas as pd
import numpy as np

ID_COL = "user_id"

# Candidate columns that might exist across apps
SCORE_COLS = [
    "compat_pct",        # okcupid
    "prob_mutual_like",  # hinge (corrected)
    "bumble_score",      # bumble
    "tinder_score",      # tinder
    "cupid_final",       # cupid
]
AUX_COLS = ["text_sim", "age", "distance_km", "lifestyle_score", "loc_score"]

def first_present(df, cols):
    for c in cols:
        if c in df.columns:
            return c
    return None

rows = []
for app, df in tables.items():
    row = {"app": app, "n": len(df)}
    primary = first_present(df, SCORE_COLS)
    row["primary_score_col"] = primary if primary else ""
    if primary:
        row["score_mean"] = float(df[primary].mean())
        row["score_std"]  = float(df[primary].std())
        row["score_p90"]  = float(df[primary].quantile(0.90))
    else:
        row["score_mean"] = row["score_std"] = row["score_p90"] = np.nan
    row["text_sim_mean"] = float(df["text_sim"].mean()) if "text_sim" in df else np.nan
    row["dist_mean_km"]  = float(df["distance_km"].mean()) if "distance_km" in df else np.nan
    row["dist_p90_km"]   = float(df["distance_km"].quantile(0.90)) if "distance_km" in df else np.nan
    row["age_mean"]      = float(df["age"].mean()) if "age" in df else np.nan
    row["age_std"]       = float(df["age"].std())  if "age" in df else np.nan
    rows.append(row)

metrics = pd.DataFrame(rows)

cols_order = [
    "app","n","primary_score_col",
    "score_mean","score_p90","text_sim_mean",
    "dist_mean_km","dist_p90_km","age_mean","age_std",
]
print("Step 5 (alt): User-centric metrics & trade-offs (fixed hinge)")
display(metrics.reindex(columns=cols_order)
        .sort_values("app")
        .reset_index(drop=True)
        .style.format({
            "score_mean":"{:.3f}","score_p90":"{:.3f}",
            "text_sim_mean":"{:.3f}",
            "dist_mean_km":"{:.1f}","dist_p90_km":"{:.1f}",
            "age_mean":"{:.1f}","age_std":"{:.1f}",
        }))
```
